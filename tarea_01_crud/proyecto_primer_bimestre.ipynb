```python
#IMPORTACIONES
import json as json
import numpy as np
```

```python
#VARIABLES GLOBALES
path="./data/base_de_datos.json"
base_de_datos={
    'estudiantes':[],
    'materias':[]
}
```

```python
#Cargar base de datos desde archivo
def obtener_base():
    global base_de_datos, path
    try:
        with open(path,encoding="utf-8") as json_file:  
            base_de_datos = json.load(json_file)          
            json_file.close()
    except:
        print("Error cargando la base de datos")
```

```python
#Guardar base de datos
def guardar_base():
    global base_de_datos, path
    try:
        with open(path,'w',encoding="utf-8") as json_file:
            json.dump(base_de_datos,json_file)
            json_file.close()
        return True
    except:
        print("Error guardando la base de datos")
        return False
```

```python
class Materia:
    codigo=''
    nombre=''
    semestre=''
    
    def __init__(self,codigo,nombre,semestre):
        self.codigo=codigo
        self.nombre=nombre
        self.semestre=semestre
        
    def __str__(self):
        return(f'Codigo: {self.codigo}\nNombre: {self.nombre}\nSemestre: {self.semestre}')
```

```python
class Estudiante:
    codigo=''
    nombre=''
    apellido=''
    facultad=''
    curriculum_academico=[]
    
    def __init__(self,codigo,nombre,apellido,facultad):
        self.codigo=codigo
        self.nombre=nombre
        self.apellido=apellido
        self.facultad=facultad
        
    def __str__(self):
        return (f'Codigo: {self.codigo}\nNombre: {self.nombre}\nApellido: {self.apellido}\nFacultad: {self.facultad}\nMaterias: {self.__obtener_curriculum()}')
    
    def agregar_materia(self,nombre_materia,nota_materia):
        registro={'nombre':nombre_materia,'nota':nota_materia}
        self.curriculum_academico.append(registro)
        
    def __obtener_curriculum(self):
        return json.dumps(self.curriculum_academico)
    
    def to_dict(self):
        estudiante={
            'codigo':self.codigo,
            'nombre':self.nombre,
            'apellido':self.apellido,
            'facultad':self.facultad,
            'curriculum_academico':[]
        }
        estudiante['curriculum_academico']=self.curriculum_academico
        return estudiante
        
```

```python
def abrir_menu_principal():
    flag_main=1
    while flag_main:
        print("-- Sistema de Adminsitración Académica --\n",
                "Seleccione una opción:\n",
                "1. Administración Estudiantes\n",
                "2. Administración Materias\n",
                "0. Salir del Sistema\n")
        opcion=input()
        if(opcion!='1' and opcion!='2' and opcion!='0'):
            print("\n***Ha ingresado un valor inválido, vuelva a intentarlo***\n")
        else:
            flag_main=0
        if(opcion=='1'):
            administrar_estudiantes()
        if(opcion=='2'):
            administrar_materias()
        if(opcion=='0'):
            print("Usted ha cerrado el aplicativo\n")
```

```python
def administrar_estudiantes():
    flag_admin_estud=1
    while flag_admin_estud:
        print("\n-- Administración de Estudiantes --\n",
                 "Seleccione una opción:\n",
                 "1. Obtener listado de estudiantes\n",
                 "2. Registrar nuevo estudiante\n",
                 "3. Consultar estudiante\n",
                 "4. Actualizar datos de estudiante\n",
                 "5. Eliminar estudiante\n",
                 "6. Ingresar nota a curriculum de estudiante\n",
                 "0. Volver\n")
        opcion=input()
        if(opcion!='1' and opcion!='2' and opcion!='3'and opcion!='4' and opcion!='5'and opcion!='6' and opcion!='0'):
            print("\n***Ha ingresado un valor inválido, vuelva a intentarlo***\n")
        if(opcion=='1'):
            obtener_estudiantes()
        if(opcion=='2'):
            registrar_estudiante()
        if(opcion=='3'):
            codigo=input("\nIngrese el código del estudiante a consultar:\n")
            consultar_estudiante(codigo)
        if(opcion=='4'):
            actualizar_estudiante()
        if(opcion=='5'):
            eliminar_estudiante()
        if(opcion=='6'):
            ingresar_materia_a_curriculum()
        if(opcion=='0'):
            abrir_menu_principal()
            flag_admin_estud=0
            
def obtener_estudiantes():
    estudiantes=base_de_datos['estudiantes']
    if(len(estudiantes)==0):
        print("\n***No se han registrado estudiantes***\n")
        return False
    else:
        print("\nLista de estudiantes:")    
        for estudiante in estudiantes:
            print("- Código: ",estudiante['codigo'],"\tNombre: ",estudiante['nombre'],"\tApellido: ",estudiante['apellido'],"\tFacultad: ",estudiante['facultad'])
        return True
    
def registrar_estudiante():
    estudiante_string=input("\nIngrese los datos del nuevo estudiante \n(Código,Nombre,Apellido,Facultad)\n")
    estudiante_array=estudiante_string.split(',')
    try:
        estudiante=Estudiante(estudiante_array[0],estudiante_array[1],estudiante_array[2],estudiante_array[3])
        index=existe_estudiante(estudiante.codigo)
        if(index==-1):
            base_de_datos['estudiantes'].append(estudiante.to_dict())
            if(guardar_base()):
                print("\n***Estudiante registrado exitosamente***\n")
            else:
                print("\n***No se pudo registrar el estudiante***\n")
        else:
            print("\n*** Error: El estudiante ya se encuentra registrado ***")
    except IndexError:
        print("\n*** Error: No se han provisto suficientes argumentos ***")
    
def existe_estudiante(codigo_estudiante):
    estudiantes=base_de_datos['estudiantes']
    index=-1
    for estudiante in estudiantes:
        if (estudiante['codigo']==codigo_estudiante):
            index=estudiantes.index(estudiante)
            break
    return index

def consultar_estudiante(codigo_estudiante):
    estudiantes=base_de_datos['estudiantes']
    index=-1
    for estudiante in estudiantes:
        if (estudiante['codigo']==codigo_estudiante):
            print("- Código: ",estudiante['codigo'],
                  "\tNombre: ",estudiante['nombre'],
                  "\tApellido: ",estudiante['apellido'],
                  "\tFacultad: ",estudiante['facultad'],
                  "\nCurriculum: ",estudiante['curriculum_academico'])
            index=estudiantes.index(estudiante)
            break
    if (index==-1):
        print("***No se encontró al estudiante en la base de datos***")
    return index
    
def eliminar_estudiante():
    codigo_estudiante=input("\nIngrese el código del estudiante que será eliminado:\n")
    index=existe_estudiante(codigo_estudiante)
    if(index!=-1):
        del base_de_datos['estudiantes'][index]
        if(guardar_base()):
            print("\n***Estudiante eliminado exitosamente***\n")
        else:
            print("\n***No se pudo eliminar el estudiante***\n")
    else:
        print("***No se encontró al estudiante en la base de datos***")
    
    
def actualizar_estudiante():
    estudiante_string=input("\nIngrese el código del estudiante a ser actualizado y sus nuevos datos\n (Código,Nuevo nombre,Nuevo apellido,Nueva facultad):\n")
    estudiante_array=estudiante_string.split(',')
    try:
        estudiante=Estudiante(estudiante_array[0],estudiante_array[1],estudiante_array[2],estudiante_array[3])
        index=existe_estudiante(estudiante.codigo)
        if(index!=-1):
            base_de_datos['estudiantes'][index]=estudiante.to_dict()
            if(guardar_base()):
                print("\n***Estudiante actualizado exitosamente***\n")
            else:
                print("\n***No se pudo actualizar el estudiante***\n")
        else:
            print("***No se encontró al estudiante en la base de datos***")
    except IndexError:
        print("\n*** Error: No se han provisto suficientes argumentos ***")
        
        
def ingresar_materia_a_curriculum():
    if(obtener_estudiantes() and obtener_materias()):
        parametros_string=input("\nIngrese los valores:\n(CódigoEstudiante,CódigoMateria,Calificación)\n")
        parametros_array=parametros_string.split(',')
        try:
            cod_estudiante=parametros_array[0]
            cod_materia=parametros_array[1]
            nota_materia=parametros_array[2]
            index_estudiante=existe_estudiante(cod_estudiante)
            if(index_estudiante!=-1):
                index_materia=existe_materia(cod_materia)
                if(existe_materia(index_materia!=-1)):
                    nombre_materia=base_de_datos['materias'][index_materia]['nombre']
                    registro={'nombre':nombre_materia,'nota':nota_materia}
                    base_de_datos['estudiantes'][index_estudiante]['curriculum_academico'].append(registro)
                    if(guardar_base()):
                        print("\n***Estudiante actualizado exitosamente***\n")
                    else:
                        print("\n***No se pudo actualizar el estudiante***\n")
                else:
                    print("***No se encontró la materia en la base de datos***")

            else:
                print("***No se encontró al estudiante en la base de datos***")
        except IndexError:
            print("\n*** Error: No se han provisto suficientes argumentos ***")
            
            

    
```
