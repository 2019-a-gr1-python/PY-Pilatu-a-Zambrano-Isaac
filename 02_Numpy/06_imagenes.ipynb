## Pixeles


Cada pixel -> valor  ! Dependiendo del tipo de imagen:

Ejemplo:

Pixel [0,0] es de color rojo, el rojo puede estar representado por el formato RGB ->  255,0,0

|0-255|0-255|0-255|
|--|--|--|
|Red|Green|Blue|

Cada *Red*, *Green* y *Blue* es conocido como un `Channel`. Por lo tanto la imagen es una imagen de *3* `Channel image`.

## Grayscale

Las imagenes en escalado de grises son representadas por una intensidad del color `negro` -> 0 - 1.

Ej:
Pixel [0,0] -> 0.2 intensidad

En este caso esta imagen seria una imagen de *1* `Channel image`

Basicamente existen dos tipos de imagenes, o de `representaciones` de imagenes. La primera es *Single Channel* y la segunda es *Multi Channel*

```python
import numpy as np
from scipy import ndimage
from scipy import misc
```

```python
mapache=misc.face()
```

```python
mapache.shape
```

```python
type(mapache)
```

```python
import matplotlib.pyplot as plot
plot.imshow(mapache)
```

```python
mapache_alterado_1=mapache[::-1]
plot.imshow(mapache_alterado_1)
```

```python
mapache_alterado_2=mapache[::2]
plot.imshow(mapache_alterado_2)
```

```python
mapache_alterado_3=mapache*2
plot.imshow(mapache_alterado_3)
```

```python
mapache_alterado_4=mapache[::10,::10]
plot.imshow(mapache_alterado_4)
```

```python
mapache_alterado_5=mapache[::,::-1]
plot.imshow(mapache_alterado_5)
```

```python
mapache_alterado_6=mapache[::,::,::-1]
plot.imshow(mapache_alterado_6)
```

```python
mapache_alterado_7=misc.face()
mapache_alterado_7.setflags(write=1)
np.random.shuffle(mapache_alterado_7)
plot.imshow(mapache_alterado_7)
```

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.1, 0.4, 0.1])

mapache_gris=rgb2gray(mapache)
plot.imshow(mapache_gris, cmap = plot.get_cmap('gray'))
```
