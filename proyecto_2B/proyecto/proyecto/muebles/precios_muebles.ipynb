```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path='./muebles.json'
df_muebles=pd.read_json(path)
```

```python
def hacer_plural(palabra_singular):
    if(palabra_singular[-1:]!='s'):
        return palabra_singular+'s'
    else:
        return palabra_singular
```

```python
df_muebles.categoria=list(map(lambda x:hacer_plural(x),df_muebles.categoria.values))
```

## Muebles mas caros


### General

```python
df_muebles.sort_values(by=['precio_regular'], ascending=False).head(10)
```

### Comedores

```python
son_comedores=df_muebles['categoria']=='Comedores'
df_muebles[son_comedores].sort_values(by=['precio_regular'], ascending=False).head(10)
```

### Dormitorios

```python
son_dormitorios=df_muebles['categoria']=='Dormitorios'
df_muebles[son_dormitorios].sort_values(by=['precio_regular'], ascending=False).head(10)
```

### Salas

```python
son_salas=df_muebles['categoria']=='Salas'
df_muebles[son_salas].sort_values(by=['precio_regular'], ascending=False).head(10)
```
